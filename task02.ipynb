# task 2.3: Identify Trends and Patterns

```python


# Scatterplot to see trend between flipper length and body mass
sns.scatterplot(x="flipper_length_mm", y="body_mass_g", hue="species", data=df)
plt.title("Flipper Length vs Body Mass")
plt.savefig("plot2.png")
plt.show()

```

```python
# Boxplot to detect anomalies 
sns.boxplot(x="species", y="bill_length_mm", data=df)
plt.title("Anomalies in Bill Length by Species")
plt.savefig("plot.png")
plt.show()
```

# task 2.4: Hypothesis Test

```python
# Hypothesis: Male penguins have higher body mass than females
sns.boxplot(x="sex", y="body_mass_g", data=df)
plt.title("Hypothesis Test: Body Mass by Sex")
plt.savefig("plot.png")
plt.show()
```
